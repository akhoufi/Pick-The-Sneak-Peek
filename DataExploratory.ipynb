```python
import pandas as pd
```

```python
metaData = pd.read_csv('movie.metadata.tsv', sep = '\t')
```

```python
metaData.columns = ['Id', 'unkwown', 'Name', 'Date', 'unkwown2', 'unknown3', 'Language', 'Country', 'Types']
```

```python
metaData
```

```python
types = metaData['Types'].as_matrix()
```

```python
types_matrix = []
for types_film in types:
    types_list = types_film.split(',')
    for i in range(len(types_list)):
        list_split = types_list[i].split(':')
        types_list[i] = list_split[len(list_split)-1]
        types_list[i] = types_list[i][2:len(types_list[i])-1]
        if i == len(types_list)-1:
            types_list[i] = types_list[i][0:len(types_list[i])-1]
    types_matrix.append(types_list)
```

```python
print(types_matrix)
```

```python
c_genres = dict()
for film in types_matrix:
    for genre in film:
        if genre in c_genres:
            c_genres[genre] = c_genres[genre] + 1
        else:
            c_genres[genre] = 1
```

```python
print(c_genres)
```

```python
import operator
dico_trie = sorted(c_genres.iteritems(), reverse=True, key=operator.itemgetter(1))
```
